# **Demo of the `StopVideo` class**
All results are saved in a folder, named as the input video, inside the /output folder. 

```python
from stop_checker import *
```

### **‚öôÔ∏è Setup new video**
Before analyzing the video, we need to define the zones and the reference point. To do this, we will extract the first frame of the video and save it in the `/first_frames` folder.

```python
get_first_frame("example")
```

Now that we have the first frame, we can visit [this link](https://roboflow.github.io/polygonzone/) to to obtain the coordinates of the corners defining the zones. These zones are later added in the `config.py`.

## **üì¶ Initalize class**
The `StopVideo` class is initialized with the name of the video saved in `/videos`.

```python
video = StopVideo("example")
```

## **üîç Perform inference**
We can now perform inference on the video.

```python
video.inference()
```

The inference results are saved as `report`. This report contains a detailed frame-by-frame analysis of the video

```python
# Example: Retrieve the report for frame 30
report30 = video.report[30]

# Example:
#{1: {'wheel': [{3: {'stopzone': True, 'outzone': False}}],
#  'speed': 802.2314192301371,
#  'car_center': (840.0, 534.0)}}

# Car with id 1 had one wheel in the frame (id 3). The wheel was in the stopzone but not in the outzone.
# The car had a speed of 802 pixels/sec and had its center coordinates at (x,y) = (840.0, 534.0)
```

## **üìä Analyze detections**
After we have performed inference, we can analyze the detections by running the `.analyze()`. This method classifies the cars.

```python
video.analyze()
```

Now we can see the result of the analysis

```python
analysis = video.analysis

# Example:
# {10: {'First Entrance': 45, 'Last Exit': 106, 'Status': 'Failed to stop'}}
# Car with id 10, was in the video from frame 45 to frame 106 and did not stop at the white line.
```
